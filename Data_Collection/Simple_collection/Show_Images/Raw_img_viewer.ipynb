# Reading in and viewing .Raw files

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

with open("depth_720.raw", "rb") as f:
    buffer = f.read()
    
depth = np.frombuffer(buffer, dtype="float32").reshape(int(720), -1)

plt.figure(figsize = (18,32)) ## make the figure larger or smaller, keep the aspect ratio (9:16) the same
plt.imshow(depth, interpolation='nearest')
plt.axis("off");
```

```python
with open("color_720.raw", "rb") as f:
    buffer = f.read()
    
color = np.frombuffer(buffer, dtype="uint8").reshape(int(720),1280 ,- 1)
plt.figure(figsize = (18,32))
plt.imshow(color)
plt.axis("off");
```

# Exploring the data structures of the images

```python
print(f"Shape of color image after reshaping is: \t {color.shape}")
print(f"Shape of depth image after reshaping is: \t {depth.shape}")
```

- Colour images are read in as 'unit8' with 4 channels, RGBA
- This means that for every one of the 720*1280 pixels there are 4 numbers that represent the Red, Green, Blue and alpha channels in the image
- The range of unit8 is 0-255

- Depth images are read in as 'float32' with 1 channel, depth
- The range of float32 is -3.4E+38 to +3.4E+38 but
- It is assumed that depth is never negative so depth will have a lower bound of 0
- It is currently unknown how the depth information relates to in game distance or real life distance
